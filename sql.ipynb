# Data processing

### covid cases

```python
covid_cases.head(11)
```

```python
#Drop unnecessary
#covid_cases.drop([ 'Unnamed: 0',], axis="columns" , inplace=True)
covid_cases=covid_cases.drop([10,21,32,43,54,65,76,87,98])
```

```python
covid_cases = covid_cases.drop_duplicates()
```

```python
# Fix the column Departm
covid_cases['departement'] = covid_cases['departement'].replace({'Guest' :'Ouest' }, inplace =False)
```

```python
#verification
covid_cases['departement'].value_counts()
```

## clean department before merge

```python
#verif
departement.head(1)
```

```python
departement=departement.loc[:,['adm1code','adm1_fr','IHSI_UNFPA_2019','IHSI_UNFPA_2019_female','IHSI_UNFPA_2019_male']]
```

```python
#Covid_cases & Department (columns merge)
covid_cases_depart = pd.merge(covid_cases,departement,how= 'left', left_on='departement', right_on='adm1_fr')
```
